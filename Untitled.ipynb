```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException
import time

# Specify the path to chromedriver.exe
chromedriver_path = 'C:/Users/gabriel.githinji/Desktop/areawork/chromedriver-win32/chromedriver.exe'

# Configure Chrome options
chrome_options = webdriver.ChromeOptions()

# Initialize the WebDriver with the specified path to chromedriver.exe and options
driver = webdriver.Chrome(service=Service(executable_path=chromedriver_path), options=chrome_options)

# URL of the webpage containing city names
url = "https://www.ubereats.com/?slr=1"

# Open the webpage
driver.get(url)

# Wait for the city links to be clickable with retry mechanism
MAX_RETRIES = 3
retry_count = 0
city_elements = None
while retry_count < MAX_RETRIES:
    try:
        city_elements = WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a[href^="/city/"]')))
        break
    except StaleElementReferenceException:
        print("StaleElementReferenceException occurred. Retrying...")
        retry_count += 1
        time.sleep(1)  # Adding a small delay before retrying

if city_elements is None:
    print("Failed to retrieve city elements after multiple retries. Exiting...")
    driver.quit()
    exit()

# Extract city names from the city URLs
city_names = [element.get_attribute("href").split("/")[-1] for element in city_elements]

# Iterate through each city
for city in city_names:
    try:
        # Construct the URL for the city
        city_url = f"https://www.ubereats.com/city/{city}"
        print(f"Visiting city: {city_url}")

        # Open the city URL with retry mechanism
        retry_count = 0
        while retry_count < MAX_RETRIES:
            try:
                driver.get(city_url)
                WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a[href^="/category/"]')))
                break
            except StaleElementReferenceException:
                print("StaleElementReferenceException occurred. Retrying...")
                retry_count += 1
                time.sleep(1)  # Adding a small delay before retrying

        if retry_count == MAX_RETRIES:
            print("Failed to load city page after multiple retries. Skipping...")
            continue

        # Extract restaurant URLs for Asian and Healthy food categories
        category_links = driver.find_elements(By.CSS_SELECTOR, 'a[href^="/category/"]')
        for link in category_links:
            category_name = link.text.strip()
            if category_name in ["Asian Food", "Healthy Food"]:
                category_url = link.get_attribute("href")
                print(f"{category_name}: {category_url}")

                # Open the category URL with retry mechanism
                retry_count = 0
                while retry_count < MAX_RETRIES:
                    try:
                        driver.get(category_url)
                        WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a[href^="/store/"]')))
                        break
                    except StaleElementReferenceException:
                        print("StaleElementReferenceException occurred. Retrying...")
                        retry_count += 1
                        time.sleep(1)  # Adding a small delay before retrying

                if retry_count == MAX_RETRIES:
                    print("Failed to load category page after multiple retries. Skipping...")
                    continue

                # Extract and print restaurant URLs
                restaurant_links = driver.find_elements(By.CSS_SELECTOR, 'a[href^="/store/"]')
                print("Restaurant Href Links:")
                for url in restaurant_links:
                    print(url.get_attribute("href"))

    except StaleElementReferenceException:
        print("StaleElementReferenceException occurred. Skipping city link.")

# Close the WebDriver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import time

# Path to the Chrome WebDriver
chromedriver_path = 'C:/Users/gabriel.githinji/Desktop/areawork/chromedriver-win32/chromedriver.exe'

# Initialize Chrome WebDriver with a Service object
service = webdriver.chrome.service.Service(chromedriver_path)
service.start()
driver = webdriver.Chrome(service=service)

# URL of the webpage containing city names
url = "https://www.ubereats.com/?slr=1"

# Open the webpage
driver.get(url)

# Wait for the city links to be clickable
city_elements = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a[href^="/city/"]'))
)

# Extract city names from the city URLs
city_names = [element.get_attribute("href").split("/")[-1] for element in city_elements]

# Iterate through each city
for city in city_names:
    try:
        # Construct the URL for the city
        city_url = f"https://www.ubereats.com/city/{city}"
        print(f"Visiting city: {city_url}")

        # Open the city URL
        driver.get(city_url)

        # Wait for category links to be clickable
        category_links = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a[href^="/category/"]'))
        )

        # Find Healthy Food category link
        healthy_food_link = next((link for link in category_links if link.text.strip() == "Healthy Food"), None)
        if healthy_food_link is None:
            print("Healthy Food category not found for this city")
            continue

        # Open Healthy Food category URL
        healthy_food_link.click()

        # Wait for restaurant links to be clickable
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a[href^="/store/"]'))
        )

        # Extract restaurant URLs
        restaurant_links = driver.find_elements(By.CSS_SELECTOR, 'a[href^="/store/"]')
        for restaurant_link in restaurant_links:
            restaurant_url = restaurant_link.get_attribute("href")

            # Open restaurant URL
            driver.get(restaurant_url)

            # Extract rating and rating count
            rating_element = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'h4[data-baseweb="typo-headingmedium"]'))
            )
            rating = float(rating_element.text)

            rating_count_element = driver.find_element(By.CSS_SELECTOR, 'p[data-baseweb="typo-paragraphxsmall"]')
            rating_count = rating_count_element.text.split()[0]

            # Extract restaurant name
            restaurant_name_element = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'h1[class*=title]'))
            )
            restaurant_name = restaurant_name_element.text.strip()

            # Ignore restaurants with rating less than 4.5
            if rating < 4.5:
                print(f"Ignoring {restaurant_name} with rating {rating} and {rating_count} ratings as it is below 4.5")
                continue

            # Print restaurant details
            print("Restaurant Name:", restaurant_name)
            print("Rating:", rating)
            print("Rating Count:", rating_count)

    except Exception as e:
        print(f"An error occurred: {e}")

# Quit the WebDriver
driver.quit()
```
